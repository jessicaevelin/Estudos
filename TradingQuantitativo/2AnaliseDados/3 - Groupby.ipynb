# Groupby

```python
import pandas as pd

data = {
    'Classe': ['Júnior','Júnior','Pleno','Pleno','Sênior','Sênior'],
    'Nome': ['Jorge', 'Carlos', 'Roberta', 'Patrícia', 'Bruno', 'Vera'],
    'Venda':[200,120,340,124,243,350]
}

df = pd.DataFrame(data)
df
```

```python
group = df.groupby('Classe') # É bom guardar em uma variável, porque pode ser muito demorado
group.min()
```

```python
df_janeiro = df.copy()
df_fevereiro = df.copy()
df_fevereiro['Venda'] = [999, 562 ,1000, 2366, 21, 0]
df_fevereiro
```

```python
df_anual = pd.concat([df_janeiro, df_fevereiro])
df_anual
```

```python
df_anual.groupby(['Classe','Nome']).sum()
```
