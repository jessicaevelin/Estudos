```python
import pandas as pd
import datetime

numeros_de_dias = 100

datas = pd.date_range(start='1/1/2021', periods=numeros_de_dias)
datas
```

```python
df = pd.DataFrame(data=range(numero_de_dias), columns=['number'], index=datas)
print(df.head(3))
print('----')
print(df.index)
```

```python
# Acessando valores do segundo lugar
df.index[1]
df.index[1].hour
df.index[1].day
df.index[1].month
df.index[1].year

# Filtrando pelo dia
print(df[df.index.day == 10])

# Selecionando o mês de toda a Serie
df['Mês'] = df.index.month
df
```

```python
# Selecionando datas condicionais
df[df.index > datetime.datetime(2021,4,1)]
```
